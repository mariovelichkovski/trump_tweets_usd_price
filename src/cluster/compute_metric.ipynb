```
TWEETS_FILE = "/content/drive/My Drive/project/dataset/preprocessed/tweets/dataset_model.csv"
TWEETS_OUTPUT_FILE = "/content/drive/My Drive/project/dataset/preprocessed/tweets/dataset_cluster.csv"
METRIC_OUTPUT_FILE = "/content/drive/My Drive/project/tmp/cluster_metric.npy"
```

```
import pandas as pd
import tensorflow_hub as hub
import numpy as np
from sklearn.metrics.pairwise import pairwise_distances
```

```
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")
```

```
tweets_np = pd.read_csv(TWEETS_FILE).values
tweets_np = tweets_np[tweets_np[:, 0] == "realDonaldTrump"]
tweets_np = tweets_np[-tweets_np[:, 3].argsort()]
tweets_np = tweets_np[:3000, :]

tweets_rdt = tweets_np[:, 2].tolist()

embeddings_rdt = embed(tweets_rdt)

df = pd.DataFrame(tweets_rdt, columns=["text"])
df.to_csv(TWEETS_OUTPUT_FILE, index=False)
```

```
def similarity_metric(a, b):
  print(a, b)
  a = embeddings_rdt[int(a[0])]
  b = embeddings_rdt[int(b[0])]
  c = np.inner(a,b)
  if c <= 0:
    return 1
  return 1 - c
```

```
X = np.arange(len(tweets_rdt)).reshape(-1, 1)
print(X.shape)
metrics_rdt = pairwise_distances(X, X, metric=similarity_metric)

np.save(METRIC_OUTPUT_FILE, metrics_rdt)
```
