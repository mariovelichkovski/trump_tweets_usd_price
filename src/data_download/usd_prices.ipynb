```
DATASET_RAW_USD_PRICES_DIRECTORY = "/content/drive/My Drive/project/dataset/raw/usd_prices"
```

```
!pip install histdata
from histdata import download_hist_data as dl
from histdata.api import Platform as P, TimeFrame as TF
```

```
!rm -rf $DATASET_RAW_USD_PRICES_DIRECTORY
%cd $DATASET_RAW_USD_PRICES_DIRECTORY

# Download data for [2017, 2019]
for year in range(2017, 2019+1):
  r = dl(
    year=str(year),
    pair="eurusd",
    platform=P.GENERIC_ASCII,
    time_frame=TF.ONE_MINUTE,
    output_directory=DATASET_RAW_USD_PRICES_DIRECTORY
  )
  !unzip '{DATASET_RAW_USD_PRICES_DIRECTORY + "/DAT_ASCII_EURUSD_M1_" + str(year) + ".zip"}'

# Download per month data for 2020
current_year = str(2020)
for month in range(1, 9+1):
  dl(
    year=current_year,
    month=str(month),
    pair="eurusd",
    platform=P.GENERIC_ASCII,
    time_frame=TF.ONE_MINUTE,
    output_directory=DATASET_RAW_USD_PRICES_DIRECTORY
  )
  month = str(month) if month >= 10 else "0"+str(month)
  !unzip '{DATASET_RAW_USD_PRICES_DIRECTORY + "/DAT_ASCII_EURUSD_M1_" + current_year + month + ".zip"}'
```
